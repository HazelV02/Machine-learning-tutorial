```python
import sklearn.metrics as metrics
def regression_results(y_true, y_pred):
    mean_absolute_error=metrics.mean_absolute_error(y_true, y_pred) 
    mse=metrics.mean_squared_error(y_true, y_pred) 
    median_absolute_error=metrics.median_absolute_error(y_true, y_pred)
    r2=metrics.r2_score(y_true, y_pred)
    print('r2: ', round(r2,4))
    print('MAE: ', round(mean_absolute_error,4))
    print('MSE: ', round(mse,4))
    print('RMSE: ', round(np.sqrt(mse),4))
```

```python
categorical_columns = ['Sex', 'Embarked']
train_obj_new = train_obj[categorical_columns]
test_obj_new = test_obj[categorical_columns]


train_obj_new = train_obj_new.fillna('Missing')
test_obj_new = test_obj_new.fillna('Missing')


onehot_encoder = OneHotEncoder(sparse_output=False)
onehot_encoder = onehot_encoder.fit(train_obj_new)


train_obj_enc = onehot_encoder.transform(train_obj_new)
test_obj_enc = onehot_encoder.transform(test_obj_new)


train_obj_enc = pd.DataFrame(train_obj_enc, columns=onehot_encoder.get_feature_names_out(categorical_columns), index=train_obj.index)
test_obj_enc = pd.DataFrame(test_obj_enc, columns=onehot_encoder.get_feature_names_out(categorical_columns), index=test_obj.index)


train_obj_drop = train_obj.drop(columns=categorical_columns)
test_obj_drop = test_obj.drop(columns=categorical_columns)


train_obj = pd.concat([train_obj_drop, train_obj_enc], axis=1)
test_obj = pd.concat([test_obj_drop, test_obj_enc], axis=1)
```
